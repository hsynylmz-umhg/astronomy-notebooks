# Çift Yıldızlar

$\qquad$ Çift yıldızlar ortak bir kütle merkezinde kütle çekim etkisiyle zıt yönlerde dolanan iki veya daha fazla yıldızdan oluşan sistemlerdir. Bu tanımlama 1802 yılında *Sir William Herschel* tarafından kullanıldı.

Çift yıldızlar, günümüzde gözlem türüne göre farklı kategorilerde sınıflandırılmaktadır:
- Optik Çift Sistemler
- Astrometrik Çift Sistemler
- Görsel Çift Sistemler
- Tayfsal Çift Sistemler
- Örten Çift Sistemler

<div id="optik-cift-sistemler"></div>
## Optik Çift Yıldızlar

$\qquad$ Gerçekte birbirinden çok uzakta olup bakış açısından kaynaklı, aynı düzlemde olduğu için çift yıldız sanılan yıldızlardır. Bu yıldızlar gerçekte ortak bir çekimsel merkeze **bağlı değildir!**

## Görsel Çift Sistemler

$\qquad$ Bileşen yıldızları, yeterli optiğe sahip teleskoplarla ayrı ayrı gözlenebilen çiftlerdir. Sistemdeki parlak yıldıza '*baş yıldız*' (veya kolaylık için '*A*' isimlendirmesi yapılır), daha sönük olana '*yoldaş yıldız*' (veya kolaylık için '*B*' isimlendirmesi yapılır) denir. Görsel çiftlerde kütle çekimsel bir bağ vardır ancak iki yıldız arasında herhangi bir madde alışverişi gerçekleşmez (gerçekleşmeme sebebi yıldızların birbirinden ayırt edilebiliyor olmaları yani birbirlerine uzak olmaları).

$\qquad$ Sistemin uzaklığı ve yoldaşın baş yıldız etrafında çizdiği yörüngesi bilinirse Kepler Yasaları vasıtasıyla

$ G(m_1 + m_2) = 4 \pi \sdot \frac{a^3}{P^2}    \xrightarrow[\text{CGS} \rightarrow \text{AUB birim dönüşümü yaparsak}]{\text{Eğer}}    m_1 + m_2 = \frac{a^3}{P^2} $

denklemi elde edilir.

<div id="tayfsal-cift-sistemler"></div>
## Tayfsal Çift Sistemler

$\qquad$ Bu sistemin bileşenleri birbirine çok yakın olduğundan ayrı ayrı gözlem yapılması mümkün olmaz ancak çiftler birbirine zıt yönde yörünge hareketi yaptığından taylafları alındığında dopler etkisininde zıt yönlerde olacağından tayf çizgileri zıt yönde kayma gösterir.

**Tayfsal Çift Yıldızlarda Yörünge Parametreleri:**<br>
$\quad$ $P$: Yörünge dönemi (gün)<br>
$\quad$ $T_0$: Enberiden geçiş zamanı (Julyen günü)<br>
$\quad$ $e$: Dış merkezlik (kesirsel)<br>
$\quad$ $\omega$: Enberi noktasının boylamı (derece)<br>
$\quad$ $\gamma$ veya $V_0$: Kütle merkezinin hızı (km/sn)<br>
$\quad$ $K_1$: Baş bileşenin dikine hız eğrisinin genliği (km/sn)<br>
$\quad$ $K_2$: Yoldaş bileşenin dikine hız eğrisinin genliği (km/sn)<br>

Bu sistemlerde bazı çiftler arasında madde alış-verişi olabilmektedir.

# Roche Modeli

$\qquad$ Sistemlerin kütle çekimsel etki alanlarını ve gaz akışlarını anlamamızı sağlayan modellemedir. Bu modelleme çiftler arasında madde akışı olan yıldızlarda "Legendre Polinomları" nın sistemlerin geometrisini tanımlamada yetersiz kalması sebebiyle yapılmaktadır. Ayrık çiftlerde küresel yapı korunacağı için minumum tedirginlik gösteren çiftler olarak kabul edilecek ve sistemlerin geometrisi ifade etmek için Legendre Polinomları yeterli olacaktır.
