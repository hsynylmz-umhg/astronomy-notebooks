# Çift Yıldızlar

$\qquad$ Çift yıldızlar ortak bir kütle merkezinde kütle çekim etkisiyle dolanan iki veya daha fazla yıldızdan oluşan sistemlerdir. Bu tanımlama 1802 yılında *Sir William Herschel* tarafından kullanıldı. Bunu yöneten temel yasa, Kepler'in 3. Yasasının Newton tarafından genelleştirilmiş halidir:
$$P^2 = \frac{4\pi^2}{G(M_1 + M_2)} a^3$$
Burada;<br/>
$P$ : Yörünge periyodu (saniye biriminden)<br/>
$a$ : İki yıldızın birbirine olan ortalama uzaklığı (yarı-büyük eksen, metre biriminden)<br/>
$M_{1,2}$ : Sıraasıyla yıldızların kütleleri (kilogram biriminden)<br>
$G$ : Evrensel çekim sabitidir ($6.674 \cdot 10^{-11} N \cdot m^2 / kg^2$)

Astronomlar; kütleleri Güneş biriminde '$M_\odot$', periyodu yıl biriminde 'yr' ve uzaklığı astronomi biriminde 'AU' alarak formülü $$P^2 (M_1 + M_2) = a^3$$ şeklinde sadeleştirir.

Çift yıldızlar, günümüzde gözlem türüne göre farklı kategorilerde sınıflandırılmaktadır:
- Optik Çift Sistemler
- Astrometrik Çift Sistemler
- Görsel Çift Sistemler
- Tayfsal Çift Sistemler

## Optik Çift Yıldızlar
$\qquad$ Gerçekte birbirinden çok uzakta olup bakış açısından kaynaklı, aynı düzlemde olduğu için çift yıldız sanılan yıldızlardır. Bu yıldızlar ortak bir çekimsel merkeze **bağlı değildir!**
